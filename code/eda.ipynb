```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
trends = pd.read_csv("../csv/trends/trends.csv")
trends.head()
```

```python
cor = trends.corr()
sns.set(rc = {'figure.figsize':(15,15)})
sns.heatmap(cor, xticklabels=cor.columns, yticklabels=cor.columns)
```

```python
cor = trends.cov()
sns.set(rc = {'figure.figsize':(15,15)})
sns.heatmap(cor, xticklabels=cor.columns, yticklabels=cor.columns)
```
